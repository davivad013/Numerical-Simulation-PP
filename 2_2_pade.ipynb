# Notebook referente ao exercício 2.2.4 - Aproximação de Padé

Queremos criar um gráfico da função $\sqrt(1+x)$ e compará-lo com as aproximações de Taylor de primeira, segunda e quarta ordem, e com as aproximações de Padé de ordem (1,1) ordem (2,2).

Determinamos as equações que queremos plotar:

### Taylor

**Primeira ordem:**
$$\sqrt{1+x} \approx 1 + \frac{x}{2}$$

**Segunda ordem:**
$$\sqrt{1+x} \approx 1 + \frac{x}{2} - \frac{x^2}{8}$$

**Quarta ordem:**
$$\sqrt{1+x} \approx 1 + \frac{x}{2} - \frac{x^2}{8} + \frac{x^3}{16} - \frac{5x^4}{128}$$

---

### Padé

**Ordem (1,1):**
$$\sqrt{1+x} \approx \frac{1 + \frac{3x}{4}}{1 + \frac{x}{4}}$$

**Ordem (2,2):**
$$\sqrt{1+x} \approx \frac{1 + \frac{3x}{4} + \frac{3x^2}{64}}{1 + \frac{x}{4} - \frac{x^2}{64}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
x = np.linspace(-0.999,2,n)

f = np.sqrt(1+x)
t1 = 1+x/2
t2 = 1+x/2-x**2/8
t4 = 1+x/2-x**2/8+x**3/16-5*x**4/128

p1 = (1+3*x/4)/(1+x/4)
p2 = (1+3*x/4+3*x**2/64)/(1+x/4-x**2/64)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, f, label='função', linewidth=3)
plt.plot(x, t1, '--', label='Taylor 1ª')
plt.plot(x, t2, '--', label='Taylor 2ª')
plt.plot(x, t4, '--', label='Taylor 4ª')
plt.plot(x, p1, '-.', label='Padé (1,1)')
plt.plot(x, p2, '-.', label='Padé (2,2)')
plt.axvline(0, )
plt.grid()
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Diferentes Aproximações de (1+x)^(1/2)')
plt.show()
```

Além disso, para ver bem a convergência de cada uma das aproximações, podemos plotar o gráfico que calcula a diferença entre a função e a aproximação.

```python
plt.figure(figsize=(12, 6))
plt.plot(x, np.abs(f-t1), label='Taylor 1ª')
plt.plot(x, np.abs(f-t2), label='Taylor 2ª')
plt.plot(x, np.abs(f-t4), label='Taylor 4ª')
plt.plot(x, np.abs(f-p1), label='Padé (1,1)')
plt.plot(x, np.abs(f-p2), label='Padé (2,2)')
plt.grid()
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Convergência das Aproximações de (1+x)^(1/2)')
plt.show()
```
